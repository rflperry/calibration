## MNIST embedding comparison

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()

# Rescale the images from [0,255] to the [0.0,1.0] range.
X_train = X_train[..., np.newaxis] / 255.0
X_test = X_test[..., np.newaxis] / 255.0

y_train = keras.utils.to_categorical(y_train)
y_test = keras.utils.to_categorical(y_test)

print("# train samples: ", len(X_train))
print("# test samples: ", len(X_test))
print('Shape: ', X_train[0].shape)
```

```python
keras.utils.to_categorical(np.unique(y_train, return_inverse=True, axis=0)[1])
```

### softmax loss

```python
# lenet_classifier = create_LeNet(out_dim=len(np.unique(y_train)))
classifier = EmbeddingNet(embed_dim=2, n_classes=len(np.unique(y_train)))
# the default learning rate of Adam might not be the best for this dataset
optimizer = tf.optimizers.Adam(1e-3)

# Training loop
n_iter = 2000
batch_size = 32
y_onehot = tf.one_hot(y_train, depth=len(np.unique(y_train)))

for i in range(1, n_iter+1):
    # Sample minibatch data indices
    minibatch_indices = np.random.choice(len(X_train), size=batch_size)

    with tf.GradientTape() as tape:
        logits = classifier(X_train[minibatch_indices])
        loss = tf.nn.softmax_cross_entropy_with_logits(
            tf.gather(y_onehot, minibatch_indices), logits)
        loss = tf.reduce_mean(loss)
        grads = tape.gradient(loss, classifier.trainable_weights)
        optimizer.apply_gradients(zip(grads, classifier.trainable_weights))

    if i % 100  == 0:
        print("Iter {:03d}: loss={:.3f}".format(i, loss))
```

```python
embeddings = classifier.get_embedding(X_test[:1000])
plot_embeddings(embeddings.numpy(), y_test[:1000])
```

```python
n_epochs = 100
batch_size = 32
validation_split = 0.33
        
def minibatch(X_train, y_train, batch_size, validation_split, validation=False):
    n_tot = X_train.shape[0]
    n_samp = int(n_tot * (1 - validation_split))
    if validation:
        n_samp = n_tot - n_samp
    for _ in range(int(n_samp / batch_size)):
        minibatch_indices = None
        while minibatch_indices is None or min(np.unique(y_train[minibatch_indices], axis=0, return_counts=True)[1]) == 1:   
            minibatch_indices = np.random.choice(n_samp, size=batch_size)
            if validation:
                minibatch_indices += (n_tot - n_samp)

        yield X_train[minibatch_indices], y_train[minibatch_indices]
        
train_gen = tf.data.Dataset.from_generator(
    lambda: minibatch(X_train[:1000], y_train[:1000], batch_size, validation_split),
    output_signature=(
        tf.TensorSpec(shape=(32, *X_train.shape[1:]), dtype=tf.float64),
        tf.TensorSpec(shape=(32, 10), dtype=tf.float32),
    ))

valid_gen = tf.data.Dataset.from_generator(
    lambda: minibatch(X_train[:1000], y_train[:1000], batch_size, validation_split, validation=True),
    output_signature=(
        tf.TensorSpec(shape=(32, *X_train.shape[1:])),
        tf.TensorSpec(shape=(32, 10)),
    ))
```

## polytope margin loss
loss is the sum of distances to the closest polytop boundary of missclassified points

```python
dim = 2
n_samples = 2000
# More noise in the moons makes the task harder
X, y = datasets.make_moons(n_samples=n_samples, shuffle=True, noise=.1)
# X, y = datasets.make_circles(n_samples=n_samples, random_state=8)
# Rescale and shift the dataset to better fit into zero-one box
X = (X + 1.6) / 4
X[:, 0] = X[:, 0] - 0.035
X[:, 1] = (X[:, 1] - 0.17) * 1.75
y_tf = tf.one_hot(y, depth=2)

model = ReluNet([100, 100])
# the default learning rate of Adam might not be the best for this dataset
optimizer = tf.optimizers.Adam()

# Training loop
n_iter = 1000
temp = 1#0.0001

for i in range(1, n_iter+1):
    with tf.GradientTape() as tape:
        logits = model(X)
#         loss = sigmoid_loss(logits, y_tf, temp=temp)
        margins = model.margin_distances(X)
        activations = model.activations(X)
        loss = margin_loss(margins, activations, y)
        grads = tape.gradient(loss, model.trainable_weights)
        optimizer.apply_gradients(zip(grads, model.trainable_weights))
    if i % 100  == 0:
        train_err = np.mean(predict_proba(logits, temp=temp) != y)
        print("Iter {:03d}: loss={:.3f} err={:.2%}".format(i, loss, train_err))
```

## Pseudo entropy loss
Loss is the negative sum of distances to the closest polytope boundary weighted by the probability of being the correct class label

```python
dim = 2
n_samples = 2000
# More noise in the moons makes the task harder
X, y = datasets.make_moons(n_samples=n_samples, shuffle=True, noise=.1)
X, y = spiral(n=n_samples)
# X, y = datasets.make_circles(n_samples=n_samples, random_state=8)
# Rescale and shift the dataset to better fit into zero-one box
X = (X + 1.6) / 4
X[:, 0] = X[:, 0] - 0.035
X[:, 1] = (X[:, 1] - 0.17) * 1.75
y_tf = tf.one_hot(y, depth=2)

model = ReluNet([10,10,10])
# the default learning rate of Adam might not be the best for this dataset
optimizer = tf.optimizers.Adam()

# Training loop
n_iter = 500
temp = 1#0.0001

for i in range(1, n_iter+1):
    with tf.GradientTape() as tape:
        logits = model(X)
#         loss = sigmoid_loss(logits, y_tf, temp=temp)
        margins = model.margin_distances(X)
        activations = model.activations(X)
        loss = pseudo_entropy_loss(margins, activations, y)
        grads = tape.gradient(loss, model.trainable_weights)
        optimizer.apply_gradients(zip(grads, model.trainable_weights))
    if i % 100  == 0:
        train_err = np.mean(predict_proba(logits, temp=temp) != y)
        print("Iter {:03d}: loss={:.3f} err={:.2%}".format(i, loss, train_err))
```
